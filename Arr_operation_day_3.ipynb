```python
import numpy as np
a = np.array([1, 2, 3, 4, 5])
b = np.array([10, 20, 30, 40, 50])

print(a + b)
print(a * b)
print(np.sqrt(a))
print(np.sum(a))
print(np.mean(b))
```
