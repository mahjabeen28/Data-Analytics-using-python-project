Practice: Employee Dataset Exploration

```python
import pandas as pd

employees = {
    "Emp_ID":[1001,1002,1003,1004,1005],
    "Name":["Aravind","Kavitha","Sanjay","Diya","Yazhini"],
    "Department":["HR","Finance","IT","HR","IT"],
    "Salary":[50000,60000,55000,58000,54000],
    "Experience":[3,4,2,4,3]
}

df = pd.DataFrame(employees)

print("Employee Dataset")
print(df)

print("\nAverage Salary")
print(df["Salary"].mean())

print("\nHighest Salary")
print(df["Salary"].max())

print("\nEmployees in HR")
print(df[df["Department"]=="HR"])

print("\nEmployees with Experience>3")
print(df[df["Experience"]>3])
```

Practice: Clean a Messy Sales Dataset

```python
import pandas as pd

data = {
    "Employee_ID": [101, 102, 103, 104, 102, 105, 106],
    "Name": ["Anand", "Bala", "Chitra", "Divya", "Bala", None, "Ezhil"],
    "Age": [25, 30, None , 22, 30, 28, None],
    "Salary": [50000, 60000, 70000, None , 60000, 45000, 55000],
}

df = pd.DataFrame(data)
print(df)

print(df.isnull())

df = (df.drop_duplicates())
print(df)

df["Age"] = df["Age"].fillna(df["Age"].mean())
print(df)

df = df.dropna(subset=["Salary"])
print(df)

df = df.rename(columns={"Name": "Employee_Name", "Age": "Employee_Age"})
print(df)
```

DAY - 7 DATA FILTERING AND TRANSFORMATION

```python
import pandas as pd

data = {
    "Region":["South","North","East","West","South","North"],
    "Product":["Laptop","Mobile","Laptop","Tablet","Mobile","Tablet"],
    "Sales":[50000,30000,45000,30000,20000,35000]
}

df = pd.DataFrame(data)

print(df)
```

Filtering

```python
south_sales = df[df["Region"] == "South"]

print(south_sales)
```

Sorting

```python
sorted_df = df.sort_values("Sales")

print(sorted_df)
```

GroupBy

```python
region_sales = df.groupby("Region")["Sales"].sum()

print(region_sales)
```

```python
product_sales = df.groupby("Product")["Sales"].sum()

print(product_sales)
```

Aggregation

```python
print(df["Sales"].sum())
```

```python
print(df["Sales"].mean())
```

```python
print(df["Sales"].max())
```

```python
print(df["Sales"].min())
```

```python
print(df["Sales"].count())
```

```python
result = df.groupby("Region")["Sales"].agg(
    ["sum","mean","max","min","count"]
)

print(result)
```

Practice: Region-wise Sales Report

```python
import pandas as pd

data = {
    'Region': ['North', 'South', 'East', 'West', 'North', 'South', 'East', 'West', 'North', 'South'],
    'Product': ['Laptop', 'Mobile', 'Laptop', 'Tablet', 'Mobile', 'Tablet', 'Laptop', 'Mobile', 'Tablet', 'Laptop'],
    'Sales_Amount': [50000, 25000, 52000, 18000, 27000, 19000, 51000, 26000, 20000, 55000],
    'Quantity': [5, 10, 4, 6, 9, 7, 5, 11, 8, 5]
}

df = pd.DataFrame(data)
print(df)

east_sales = df[df["Region"] == "East"]
print(east_sales)

sorted_df = df.sort_values("Sales_Amount")
print(sorted_df)

region_sales = df.groupby("Region")["Sales_Amount"].sum()
print(region_sales)

result = df.groupby("Region")["Sales_Amount"].agg(
    ["sum","mean","max","min","count"]
)
print(result)
```

DAY - 8 Data Visualization Basics (Matplotlip,Line,Bar,pie chart and Dashboard practice)

Line Chart

```python
import matplotlib.pyplot as plt

months = ["Jan", "Feb", "Mar", "Apr", "May"]

sales = [10000,15000,18000,22000,25000]

plt.plot(months, sales)

plt.title("Monthly Sales")

plt.xlabel("Months")

plt.ylabel("Sales Amount")

plt.show()
```

```python
import matplotlib.pyplot as plt

months = ["Jan", "Feb", "Mar", "Apr", "May"]

sales = [10000,15000,18000,22000,25000]

plt.plot(months, sales,
         color="green",
         marker="o",
         linestyle="--")

plt.title("Monthly Sales")

plt.xlabel("Months")

plt.ylabel("Sales Amount")

plt.show()
```

Bar Chart

```python
import matplotlib.pyplot as plt

students = ["A","B","C","D","E"]

marks = [75,80,68,89,90]

plt.bar(students,marks)

plt.title("Student Marks")

plt.xlabel("Students")

plt.ylabel("Marks")

plt.show()
```

```python
import matplotlib.pyplot as plt

students = ["A","B","C","D","E"]

marks = [75,80,68,89,90]

plt.bar(students,
        marks,
        color = ("Red","Blue","Green","Yellow","Pink"))

plt.title("Student Marks")

plt.xlabel("Students")

plt.ylabel("Marks")

plt.show()
```

Pie Chart

```python
import matplotlib.pyplot as plt

expenses = [6000,2000,4000]

categories = ["Food","Travel","Shopping"]

plt.pie(expenses,
        labels = categories,
        autopct = "%1.1f%%", colors = ["green","Pink","Yellow"])

plt.show()
```

Dashboard

```python
import matplotlib.pyplot as plt

months = ["Jan","Feb","Mar","Apr","May","Jun"]

sales = [10000,12000,15000,13000,18000,20000]

plt.figure(figsize=(18,5))

#Line Chart
plt.subplot(1,3,1)

plt.plot(months,
         sales,
         marker = "o")

plt.title("Sales Trend")

plt.xlabel("Months")

plt.ylabel("Sales")

#Bar Chart
plt.subplot(1,3,2)

plt.bar(months,
         sales,
         color = "orange")

plt.title("Sales Comparison")

plt.xlabel("Months")

plt.ylabel("Sales")

#Pie Chart
plt.subplot(1,3,3)

expenses = [6000,2000,4000]

categories = ["Food","Travel","Shopping"]

plt.pie(expenses,
        labels = categories,
        autopct = "%1.1f%%", colors = ["green","Pink","Yellow"])

plt.title("Expenses Distribution")

plt.tight_layout()

plt.show()
```
