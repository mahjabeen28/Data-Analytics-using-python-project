```python
marks = [85, 92, 78, 90, 65]

total_marks = sum(marks)
average = total_marks / len(marks)

print(f"Total Marks: {total_marks}")
print(f"Average Mark: {average:.2f}")

if average >= 90:
    grade = "A Grade (Excellent)"
elif average >= 80:
    grade = "B Grade (Very Good)"
elif average >= 70:
    grade = "C Grade (Good)"
elif average >= 50:
    grade = "D Grade (Pass)"
else:
    grade = "F Grade (Fail)"

print(f"Final Grade: {grade}")
```
